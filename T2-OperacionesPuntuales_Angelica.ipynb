## Ajuste de contraste (normalización)

```
def normalize(img):
    minimo = np.min(img)
    maximo = np.max(img)
    norm = (img - minimo)/(maximo - minimo)
    return norm
```

```
figura_6_norm = normalize(figura_6)
figura_7_norm = normalize(figura_7)
figura_8_norm = normalize(figura_8)
```

# Función submatriz

Como no hubo restricciones generales, esta función funciona siempre que se puede llevar a cabo la operación, es decir, cuando k es menor que el tamaño de la matriz y el desplazamiento en $x$ y $y$ quede dentro de los limites de la matriz.


```
def SubMatriz(img, coordenadas, k):
    """
    Recibe:
        img: imagen de cve que se puede tratar con numpy.
        coordenadas: tupla x,y que es el centro de la submatriz
        k: tamaño en x & en y de la submatriz. Siempre impar.
    Regresa: submatriz de tamaño kxk con centro en coordenadas.
    """
    desplazamiento = k//2 #es el numero de unidades que se tendrá que desplazar las coordenadas para crear las esquinas de la nueva submatriz
    superior_izquierdo = (coordenadas[0]-desplazamiento, coordenadas[1]-desplazamiento) #la coordenada de la esquina superior izquierda

    sub_matriz =[] #donde iré guardando la nueva imagen

    for i in range(k): #por cada fila obtengo las columnas del rango (todo lo contenido en x)
        sub_matriz.append( img[superior_izquierdo[1]+i][superior_izquierdo[0]:superior_izquierdo[0]+k])
    
    return np.array(sub_matriz)
```

# Filtros máximo y mínimo

El tamaño del kernel de entrada es siempre impar

## Máximo

```
def maximumBoxFilter(n, img):
    """
    Recibe:
        n: tamaño del kernel
        img: imagen a la cual poner el filtro
    """
    #Creo el padding
    color = [0,0,0]
    k = n//2 +1
    constant= cv2.copyMakeBorder(img.copy(),k,k,k,k,cv2.BORDER_CONSTANT,value=color)

    #obtener el tamaño de la imagen original
    tamanio = img.shape #está presentado como (y, x)

    #donde guardaré la nueva imagen
    nueva_imagen = []

    for i in range(k, tamanio[0]+k): #por cada fila
        fila = []
        for j in range(k, tamanio[1]+k): #por cada columna
            sub_matriz = SubMatriz(constant, (j, i), n) #obtengo la submatriz
            fila.append(np.max(sub_matriz)) #obtengo su valor maximo

        nueva_imagen.append(np.array(fila)) #meto la fila a la imagen

    return np.array(nueva_imagen) #regreso una matriz de numpy
```

```
plt.figure()
plt.imshow(figura_6_norm, 'gray'),plt.title('Normalizada')
plt.figure()
plt.imshow(maximumBoxFilter(3, figura_6_norm), 'gray'),plt.title('Filtro de Maximo')
```

Por la forma en la que es usada, mejora al tener mayor visibilidad de venas que aparecen en la tomografía.

```
plt.figure()
plt.imshow(figura_7_norm, 'gray'),plt.title('Normalizada')
plt.figure()
plt.imshow(maximumBoxFilter(3, figura_7_norm), 'gray'),plt.title('Filtro de Maximo')
```

Como la imagen es muy blanca esta imagen empeora

Al ser una imagen muy oscura mejora al poder ver mas detalles.

## Mínimo

```
def minimunBoxFilter(n, img):
    """
    Recibe:
        n: tamaño del kernel
        img: imagen a la cual poner el filtro
    """
    #Creo el padding
    color = [255, 255, 255]
    k = n//2 +1
    constant= cv2.copyMakeBorder(img.copy(),k,k,k,k,cv2.BORDER_CONSTANT,value=color)

    #obtener el tamaño de la imagen original
    tamanio = img.shape #está presentado como (y, x)

    #donde guardaré la nueva imagen
    nueva_imagen = []

    for i in range(k, tamanio[0]+k): #por cada fila
        fila = []
        for j in range(k, tamanio[1]+k): #por cada columna
            sub_matriz = SubMatriz(constant, (j, i), n) #obtengo la submatriz
            fila.append(np.min(sub_matriz)) #obtengo su valor minimo

        nueva_imagen.append(np.array(fila)) #meto la fila a la imagen

    return np.array(nueva_imagen) #regreso una matriz de numpy
```

```
plt.figure()
plt.imshow(figura_6_norm, 'gray'),plt.title('Normalizada')
plt.figure()
plt.imshow(minimunBoxFilter(9, figura_6_norm), 'gray'),plt.title('Filtro de Minimo')
```

Empeora ya que se pierde información por ser una imagen con presencia mayoritaria de negro, independientemente del tamaño del kernel.

```
plt.figure()
plt.imshow(figura_7_norm, 'gray'),plt.title('Normalizada')
plt.figure()
plt.imshow(minimunBoxFilter(3, figura_7_norm), 'gray'),plt.title('Filtro de Minimo')
```

Al ser una imagen en su mayoría blanca mejora al resaltar los detalles oscuros.

```
plt.figure()
plt.imshow(figura_8_norm, 'gray'),plt.title('Normalizada')
plt.figure()
plt.imshow(minimunBoxFilter(3, figura_8_norm), 'gray'),plt.title('Filtro de Minimo')
```

Empeora al ser una imagen oscura se oscurece mas y se pierde información.
